```python
import pandas as pd
```

```python
monthly_data = pd.read_excel('Economic Data.xlsm', sheet_name='Monthly Data')
quarterly_data = pd.read_excel('Economic Data.xlsm', sheet_name='Quarterly Data')
```

#### **Converting Monthly Data (Monthly Data.xlsm) to Quarterly Data**

```python
quarterly_avg = pd.DataFrame()

for column in monthly_data.columns:
    if column != 'Date':
        avg_temp = monthly_data.resample('Q', on='Date')[column].mean()
        
        quarterly_avg[column] = avg_temp

quarterly_avg = quarterly_avg.reset_index()

quarterly_avg['Fiscal Year'] = quarterly_avg['Date'].dt.year
quarterly_avg['Fiscal Quarter'] = quarterly_avg['Date'].dt.quarter
```

```python
quarterly_data['Fiscal Year'] = quarterly_data['Date'].dt.year
quarterly_data['Fiscal Quarter'] = quarterly_data['Date'].dt.quarter
```

```python
quarterly_financial_data = pd.read_excel('Quarterly Financial Data.xlsx')
```

```python
quarterly_financial_data = quarterly_financial_data.dropna(subset=['Fiscal Quarter'])
quarterly_financial_data['Fiscal Quarter'] = quarterly_financial_data['Fiscal Quarter'].astype('Int64')
```

```python
merged1 = pd.merge(quarterly_financial_data, quarterly_avg, how='left', on=['Fiscal Year', 'Fiscal Quarter'])
merged_quarterly_financial_data_quarterly_avg = pd.merge(merged1, quarterly_data, how='left', on=['Fiscal Year', 'Fiscal Quarter'])
```

```python
merged_quarterly_financial_data_quarterly_avg
```

```python
merged_quarterly_financial_data_quarterly_avg.to_excel('Quarterly Financial Data (Merged).xlsx')
```

#### **Converting Monthly Data (Monthly Data.xlsm) to Annual Data**

```python
annual_avg = pd.DataFrame()

for column in monthly_data.columns:
    if column != 'Date':
        annual = monthly_data.resample('Y', on='Date')[column].mean()
        
        annual_avg[column] = annual

for column in quarterly_data.columns:
    if column != 'Date':
        annual = quarterly_data.resample('Y', on='Date')[column].mean()
        
        annual_avg[column] = annual

annual_avg = annual_avg.reset_index()

annual_avg['Data Year - Fiscal'] = annual_avg['Date'].dt.year
```

```python
annual_avg = annual_avg.drop(columns={'Fiscal Year'})
annual_avg = annual_avg.iloc[:-1, :]
annual_avg
```

```python
annual_financial_data = pd.read_excel('Annual Financial Data.xlsx')
```

```python
annual_financial_data = annual_financial_data.dropna(subset=['Data Year - Fiscal'])
annual_financial_data['Data Year - Fiscal'] = annual_financial_data['Data Year - Fiscal'].astype('Int64')
```

```python
merged_annual_financial_data_annual_avg = pd.merge(annual_financial_data, annual_avg, how='left', on=['Data Year - Fiscal'])
```

```python
merged_annual_financial_data_annual_avg = merged_annual_financial_data_annual_avg.drop(columns={'Date_y'})
merged_annual_financial_data_annual_avg = merged_annual_financial_data_annual_avg.rename(columns={'Date_x': 'Date'})
```

```python
merged_annual_financial_data_annual_avg.to_excel('Annual Financial Data (Merged).xlsx')
```
